```python
from PIL import Image
import os
from deepface import DeepFace
from deepface.detectors import FaceDetector
import cv2

# Path to the input image
img_path = "biob.jpg"

# Choose the face detector model ("retinaface" in this case)
detector_name = "retinaface"

# Load the input image using OpenCV
img = cv2.imread(img_path)

# Create a named window to display the processed image
cv2.namedWindow("Faces", cv2.WINDOW_NORMAL)

# Build the face detector using the specified model
detector = FaceDetector.build_model(detector_name)

# Initialize a counter for detected faces
count = 0

# Detect faces using the chosen detector
obj = FaceDetector.detect_faces(detector, detector_name, img)

# Create a folder to save cropped images
save_folder = 'cropped_images12'
os.makedirs(save_folder, exist_ok=True)

# List to store recognition results
rec = []
scale_factor = 1.5 
# Loop through each detected face
for face in obj:
    bbox = face[1]
    x, y, width, height = bbox
    center_x = x + width / 2
    center_y = y + height / 2
    
    # Calculate new dimensions of the bounding box (adjusting for scale_factor)
    new_width = int(width * scale_factor)
    new_height = int(height * scale_factor)
    
    # Calculate new bounding box coordinates
    new_x = max(0, int(center_x - new_width / 2))
    new_y = max(0, int(center_y - new_height / 2))
    
    # Ensure the new bounding box stays within image boundaries
    new_x = min(new_x, img.shape[1] - new_width)
    new_y = min(new_y, img.shape[0] - new_height)
    
    # Crop the face with adjusted dimensions
    cropped_image = img[new_y:new_y + new_height, new_x:new_x + new_width]
    
    # Convert cropped face to a PIL image
    pil_image = Image.fromarray(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    
    # Save the cropped image with a unique filename
    cropped_image_filename = f"cropped_image_0.jpg"
    pil_image.save(os.path.join(save_folder, cropped_image_filename))
    
    # Perform face recognition on the cropped image
    result = DeepFace.find(img_path=cropped_image_filename, db_path='genimg', model_name="Facenet512", enforce_detection="False", detector_backend='retinaface', distance_metric="euclidean_l2")
    rec.append(result)

    # Increment the counter for detected faces
    count += 1

# Initialize a set to store unique recognized identities
unique_tidens = set()

# Loop through the recognition results and extract identities
for i, result_list in enumerate(rec):
    if len(result_list) > 0 and not result_list[0].empty:
        print(f"Recognized identities for frame {i}:")
        for result_df in result_list:
            identities = result_df['identity']
            for identity in identities:
                tiden = identity.split('\\')[-1].split('/')[-2]  # Extract 'tiden' part from the file path
                unique_tidens.add(tiden)
        for tiden in unique_tidens:
            name = tiden
    else:
        name = "unknown"

    # Draw rectangle and label on the original image
    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 2)
    cv2.putText(img, name, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.50, (0, 0, 255), 2)

# Display the modified image with rectangles and labels
cv2.imshow("Faces", img)

# Wait for a key press and close windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from PIL import Image
import os
from deepface import DeepFace
import cv2

detector_name = "retinaface"

cv2.namedWindow("Faces", cv2.WINDOW_NORMAL)

detector = FaceDetector.build_model(detector_name)  # Set opencv, ssd, dlib, mtcnn, or retinaface

camera = cv2.VideoCapture(0)  # Use '0' for default camera, or specify the camera index if you have multiple cameras

while True:
    ret, frame = camera.read()
    if not ret:
        break

    # Detect faces using the detector
    obj = FaceDetector.detect_faces(detector, detector_name, frame)
    
    # Initialize the list to store identities
    identities = []

    for face in obj:
        bbox = face[1]
        x, y, width, height = bbox
        x1 = x - 20
        y1 = y - 20
        cropped_image = frame[y1:y1 + height + 50, x1:x1 + width + 50]
        pil_image = Image.fromarray(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))

        # Use a unique filename for each cropped face
        cropped_filename = f"cropped_image_{x}_{y}.jpg"
        pil_image.save(cropped_filename)

        # Perform face recognition for each cropped face
        result = DeepFace.find(img_path=cropped_filename,
                               db_path='genimg',
                               model_name="Facenet512",
                               enforce_detection=False,
                               detector_backend='retinaface')

        # Store the identity in the list
        if len(result) > 0 and not result[0].empty:
            identities.extend(result[0]['identity'])
        else:
            identities.append("unknown")

        cv2.rectangle(frame, (x, y), (x + width, y + height), (0, 255, 0), 2)

    # Draw recognized identities for all detected faces
    for i, face in enumerate(obj):
        bbox = face[1]
        x, y, _, _ = bbox
        name = identities[i]
        cv2.putText(frame, name, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.50, (0, 0, 255), 2)

    cv2.imshow("Faces", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit the application
        break

camera.release()
cv2.destroyAllWindows()
```

```python
from PIL import Image
import os
from deepface import DeepFace
from deepface.detectors import FaceDetector
import cv2

# Function to capture an image from the camera
def capture_image():
    cap = cv2.VideoCapture(0)  # 0 for default camera, change the number if you have multiple cameras
    ret, frame = cap.read()
    cap.release()
    return frame

detector_name = "retinaface"

# Capture an image from the camera
img = capture_image()

cv2.namedWindow("Faces", cv2.WINDOW_NORMAL)

detector = FaceDetector.build_model(detector_name)  # Set opencv, ssd, dlib, mtcnn, or retinaface

# Detect faces using the detector
obj = FaceDetector.detect_faces(detector, detector_name, img)
save_folder = 'cropped_images12'
os.makedirs(save_folder, exist_ok=True)

# Initialize the list to store identities
identities = []
scale_factor = 1.5  # You can adjust this factor to control the size of the cropped face

for face in obj:
    bbox = face[1]
    x, y, width, height = bbox
    
    # Calculate dimensions of the cropped region based on the bounding box
    center_x = x + width / 2
    center_y = y + height / 2
    new_width = int(width * scale_factor)
    new_height = int(height * scale_factor)
    
    # Calculate new bounding box coordinates
    new_x = max(0, int(center_x - new_width / 2))
    new_y = max(0, int(center_y - new_height / 2))
    
    # Ensure that the new bounding box stays within image boundaries
    new_x = min(new_x, img.shape[1] - new_width)
    new_y = min(new_y, img.shape[0] - new_height)
    
    # Crop the face with adjusted dimensions
    cropped_image = img[new_y:new_y + new_height, new_x:new_x + new_width]
    pil_image = Image.fromarray(cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB))
    pil_image.save("cropped_image_1.jpg")
    rec=[]
    result = DeepFace.find(img_path = 'cropped_image_1.jpg', db_path = 'genimg',model_name="Facenet512", enforce_detection="False", detector_backend='retinaface',distance_metric="euclidean_l2")
    rec.append(result)
    unique_tidens = set()
    for i, result_list in enumerate(rec):
        if len(result_list) > 0 and not result_list[0].empty:
            print(f"Recognized identities for frame {i}:")
            for result_df in result_list:
                identities = result_df['identity']
                for identity in identities:
                    tiden = identity.split('\\')[-1].split('/')[-2]  # Extract 'tiden' part from the file path
                    unique_tidens.add(tiden)
                for tiden in unique_tidens:
                    name=tiden
                
        else:
            name="unknown"

    cv2.rectangle(img, (x, y), (x + width, y + height), (0, 255, 0), 2)

    cv2.putText(img, name, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.50, (0, 0, 255), 2)

cv2.imshow("Faces", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
                        
```
